```python
features = ['pctWInvInc','pctWPubAsst','PctPopUnderPov','PctUnemployed','TotalPctDiv','PctFam2Par','PctKids2Par','PctYoungKids2Par','PctTeen2Par','PctPersOwnOccup']
X = df.loc[:, features]
y = df.loc[:, ['ViolentCrimesPerPop']]
```
